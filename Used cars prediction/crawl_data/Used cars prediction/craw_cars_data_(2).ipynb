# Import libraries


```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
import time
import random
import pandas as pd
import numpy as np
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from unidecode import unidecode
```

```python
cars_dataframe = pd.read_csv('car_data.csv')
cars_dataframe.info()
```

```python
# Khởi tạo Edge WebDriver
edge_driver_path = "C:/Users/Admin/Downloads/edgedriver_win64/msedgedriver.exe"
options = Options()
service = Service(executable_path=edge_driver_path)
driver = webdriver.Edge(service=service, options=options)

# Tạo danh sách để lưu cả link, tiêu đề và các thông tin khác
car_data_with_prices = []
for index, row in cars_dataframe.iterrows():
    link = row['Link']
    title = row['Title']

    # Truy cập từng liên kết
    driver.get(link)

    try:
        # Đợi cho phần tử list-info xuất hiện
        list_info_element = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.CSS_SELECTOR, '.list-info'))
        )

        # Lấy tất cả các phần tử thông tin
        info_items = list_info_element.find_elements(By.TAG_NAME, 'li')

        car_info = {'Link': link, 'Title': title}

        for item in info_items:
            label = item.find_element(By.TAG_NAME, 'label').text.strip()

            # Kiểm tra nếu có div nhỏ để lấy giá trị
            try:
                value = item.find_element(By.XPATH, './div').text.strip() if item.find_elements(By.XPATH, './div') else item.text.split(label)[1].strip()
            except Exception:
                value = None  # Gán None nếu không tìm thấy

            car_info[label] = value

        car_data_with_prices.append(car_info)

    except Exception as e:
        print(f"Không thể lấy dữ liệu cho {link}: {e}")
        car_data_with_prices.append({'Link': link, 'Title': title, 'Error': str(e)})

# Tạo DataFrame từ dữ liệu đã thu thập
cars_dataframe_with_prices = pd.DataFrame(car_data_with_prices)
```
