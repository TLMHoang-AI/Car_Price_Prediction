# Import libraries

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
import time
import random
import pandas as pd
import numpy as np
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
```

# Crawling

## Load datas

```python
edge_driver_path = "C:/Users/Admin/Downloads/edgedriver_win64/msedgedriver.exe"

# Cấu hình tùy chọn cho Edge
options = Options()

# Khởi tạo Service cho Edge WebDriver
service = Service(executable_path=edge_driver_path)

# Tạo đối tượng trình duyệt Edge
driver = webdriver.Edge(service=service, options=options)

# Mở trang web cần cào
driver.get("https://oto.com.vn/mua-ban-xe")

# Đợi trang tải (đợi ngẫu nhiên từ 10 đến 15 giây)
time.sleep(random.randint(10, 15))
count = 1
# Lặp qua việc nhấn nút "Hiển thị thêm" cho đến khi không còn nút
while True:
    try:
        # Tìm nút "Hiển thị thêm" dựa trên class
        load_more_button = driver.find_element(By.XPATH, "//span[contains(@class, 'btn-loadmore')]")
        
        # Kiểm tra xem nút có hiển thị không và có thể nhấn không
        if load_more_button.is_displayed():
            # Nhấn vào nút
            load_more_button.click()
            print(f"Đã nhấn vào nút 'Hiển thị thêm' lần {count}")
        
        time.sleep(count + random.randint(20, 25))
        count += 1
        
    except:
        # Nếu không tìm thấy nút hoặc có lỗi, thoát khỏi vòng lặp
        print("Không còn nút 'Hiển thị thêm' hoặc không thể nhấn.")
        break
```

## Get data

```python
elems = driver.find_elements(By.CSS_SELECTOR, '.title [href]')

# Tạo danh sách để lưu cả link và tiêu đề (title)
car_data = []

# Lặp qua các phần tử để lấy link và tiêu đề
for elem in elems:
    link = elem.get_attribute('href')  # Lấy link
    
    # Kiểm tra nếu link bắt đầu bằng 'https://oto.com.vn/mua-ban-xe'
    if link.startswith('https://oto.com.vn/mua-ban-xe'):
        title = elem.text  # Lấy tiêu đề từ phần tử
        car_data.append((link, title))  # Lưu cả link và tiêu đề vào danh sách

# In ra link và tiêu đề tương ứng
for link, title in car_data:
    print(f'Tiêu đề: {title} - Link: {link}')
```

```python
for i in range(len(car_data)):
    link, title = car_data[i]

    parts = title.split(" - ")

    if len(parts) >= 2:
        truncated_title = " - ".join(parts[:2])
        car_data[i] = (link, truncated_title)

for link, truncated_title in car_data:
    print(f'Tiêu đề đã cắt: {truncated_title} - Link: {link}')

cars_dataframe = pd.DataFrame(car_data, columns=['Link', 'Title'])
```
