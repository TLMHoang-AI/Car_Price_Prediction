# Import libraries

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
import time
import random
import pandas as pd
import numpy as np
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from fuzzywuzzy import fuzz
```

# Crawling

```python
edge_driver_path = "C:/Users/Admin/Downloads/edgedriver_win64/msedgedriver.exe"

options = Options()

service = Service(executable_path=edge_driver_path)

driver = webdriver.Edge(service=service, options=options)
```

```python

driver.get("https://xebiz.vn/gia-xe/")

vehicle_data = []
while True:
    vehicle_items = WebDriverWait(driver, 10).until(
        EC.visibility_of_all_elements_located((By.CSS_SELECTOR, ".wpcm-listings-item"))
    )

    for item in vehicle_items:
        link = item.find_element(By.CSS_SELECTOR, "a").get_attribute("href")
        name = item.find_element(By.CSS_SELECTOR, ".wpcm-title").text
        
        vehicle_data.append({
            'name': name,
            'link': link
        })

    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.CSS_SELECTOR, "a.next.page-numbers"))
        )
        if 'disabled' not in next_button.get_attribute("class"):
            next_button.click()
            WebDriverWait(driver, 10).until(EC.staleness_of(next_button))
        else:
            break
    except Exception as e:
        print("No next button found or an error occurred:", e)
        break

```

```python
for vehicle in vehicle_data:
    print(f"Name: {vehicle['name']}, Link: {vehicle['link']}")
```

# Save to CSV

```python
vehicle_data.to_csv('new_cars_data.csv', index=False, encoding='utf-8')
```
